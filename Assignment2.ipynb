# Salt and Pepper Noise

```python
#Salt and Pepper
from skimage.util import random_noise
import numpy as np
def sp():
    noise_img=random_noise(gray,mode='s&p',amount=0.1)
    noise_img=np.array(255*noise_img,dtype='uint8')
    cv2.imwrite('sp_42049.png',noise_img)
#     cv2.imshow('blur',noise_img)
#     cv2.waitKey(1000)
#     cv2.destroyAllWindows()
    return noise_img

def find_avg(n):
    li1=[]
    for i in range(n):
        li1.append(sp());
    avg_img=np.mean(li1,axis=0)
    avg_img=avg_img.astype(np.uint8)
    return avg_img;
img=[]
img.append(find_avg(5))
img.append(find_avg(10))
img.append(find_avg(15))
img.append(find_avg(20))
img.append(find_avg(25))
img.append(find_avg(30))
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(img[i],'gray')
```

# Gaussion Noise

```python
#Gaussian


def noise():
    noise2_img = random_noise(gray,mode='gaussian',mean=0,var=0.01)
    noise2_img = np.array(255*noise2_img, dtype = 'uint8')
    return noise2_img
def add_to_li(n):
    li=[]
    for i in range(n):
        li.append(noise())
    avg_img=np.mean(li,axis=0)
    avg_img=avg_img.astype(np.uint8)
    return avg_img;
img=[]
#     plt.imshow(avg_img)
#     plt.show()
#     plt.subplot(2,3,i+1),plt.imshow(avg_img,'noised')
# def show(li):
#     avg_img=np.mean(li,axis=0)
#     avg_img=avg_img.astype(np.uint8)
#     plt.imshow(avg_img)
#     plt.show()
    
img.append(add_to_li(5));
img.append(add_to_li(10));
img.append(add_to_li(15));
img.append(add_to_li(20));
img.append(add_to_li(25));
img.append(add_to_li(30));
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(img[i],'gray')
```

# Speckle Noise

```python
#Speckle Noise
# gauss = np.random.normal(0,1,gray.size)
# gauss = gauss.reshape(gray.shape[0],gray.shape[1],gray.shape[2]).astype('uint8')
# noise = gray + gray * gauss
def speckle():
    row,col = gray.shape
    gauss = np.random.randn(row,col)
#     print(gauss)
    gauss = gauss.reshape(row,col)        
    noisy = gray + gray * gauss
    return noisy
def avg_func(n):
    li=[]
    for i in range(n):
        li.append(speckle())
    avg_img=np.mean(li,axis=0)
    avg_img=avg_img.astype(np.uint8)
    return avg_img;
img=[]
img.append(avg_func(5))
img.append(avg_func(10))
img.append(avg_func(15))
img.append(avg_func(20))
img.append(avg_func(25))
img.append(avg_func(30))
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(img[i],'gray')

# plt.imshow(noisy,'gray')
```
